```python
bus_ridership['Month'] = bus_ridership['Month_Beginning'].dt.month
bus_ridership['Year'] = bus_ridership['Month_Beginning'].dt.year
```

```python
bus_ridership['avg_mo_total'] = 4.34 * (5 * bus_ridership['Avg_Weekday_Rides'] + bus_ridership['Avg_Saturday_Rides'] + bus_ridership['Avg_Sunday-Holiday_Rides'])
```

```python
bus_ridership.head()
```

```python
ridership_by_year = bus_ridership.groupby(['route_stripped', 'Year'])['route', 'avg_mo_total'].sum().unstack(fill_value=0).stack().reset_index()
```

```python
ridership_by_year.head()
```

```python
bus_route_merge_data = bus_ridership[['routename', 'route_stripped', 'Year']].drop_duplicates()
```

```python
bus_route_merge_data.head()
```

```python
bus_route_merge_data.shape
```

```python
# bus_route_merge_data[bus_route_merge_data['route_stripped'] == '1']
```

```python
annual_ridership_by_line = pd.merge(left=ridership_by_year, right=bus_route_merge_data, how='left', left_on=['route_stripped', 'Year'], right_on=['route_stripped', 'Year'])
```

```python
annual_ridership_by_line.head()
```

```python
bus_merge = pd.merge(left=cta_bus_routes, right=annual_ridership_by_line, how='left', left_on='ROUTE_STRIPPED', right_on='route_stripped')
bus_merge.drop_duplicates(['ROUTE_STRIPPED', 'Year', 'avg_mo_total'], inplace=True)
```

```python
lw = 3
base = chicago_city.plot(color='white', edgecolor='black', figsize=(15,15))
base.axis('off')
year_df = bus_merge[bus_merge['Year'] == 2016]
year_df[(year_df['avg_mo_total'] > 5000000) & (year_df['avg_mo_total'] <= np.inf)].plot(linewidth=lw, color='#1a9641', alpha=0.6, ax=base, label='Over 5 million annual riders')
year_df[(year_df['avg_mo_total'] > 3000000) & (year_df['avg_mo_total'] <= 5000000)].plot(linewidth=lw, color='#a6d96a', alpha=0.6, ax=base, label='Between 3 and 5 million annual riders')
year_df[(year_df['avg_mo_total'] > 1000000) & (year_df['avg_mo_total'] <= 3000000)].plot(linewidth=lw, color='#fdae61', alpha=0.6, ax=base, label='Between 1 and 3 million annual riders')
year_df[(year_df['avg_mo_total'] > 500000) & (year_df['avg_mo_total'] <= 1000000)].plot(linewidth=lw, color='#d7191c', alpha=0.6, ax=base, label='Between 0.5 and 1 million annual riders')
base.legend(fontsize=14)
plt.tight_layout()
```
